```python
import pandas as pd

# load datasets
df_train = pd.read_csv('../../../data/processed/ohe_25K_tracks_features_and_labels_for_training.csv')
df_test = pd.read_csv('../../../data/processed/ohe_25K_tracks_features_and_labels_for_test.csv')

# don't need to seperate them since we will use pyCaret
df_union = pd.concat([df_train, df_test], ignore_index=True)
```

```python
df_union.columns
```

```python
# define features and label
features = [feature for feature in df_union.columns if feature not in ['Unnamed: 0', 'track_id', 'depth_1_genre_name', 'depth_2_genre_name',
       'depth_3_genre_name', 'depth_4_genre_name', 'depth_5_genre_name', 'track_genre_top',
       'artist_name', 'track_title', 'set_split', 'set_subset']]
label = 'track_genre_top'
```

```python
df_union['track_genre_top'].value_counts()
```

```python

# to check if the dataset is balanced.
import matplotlib.pyplot as plt

def plot_value_counts(ax, column_name, df):
    value_counts = df[column_name].value_counts()
    ax.bar(value_counts.index, value_counts.values, color='skyblue')
    ax.set_title(f'Distribution of {column_name}')
    ax.set_xlabel('Class')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better visibility

columns_to_plot = ['track_genre_top']

# Create a figure and a 5x1 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 6))

for ax, column in zip(axes, columns_to_plot):
    plot_value_counts(ax, column, df_union)

plt.tight_layout()  # Adjust layout for better visibility
plt.show()
```

```python
# EDA
# distribution for features
g = sns.FacetGrid(df_union, col='track_genre_top', col_wrap=4, height=3)
g.map(plt.hist, "tempo", bins=20, color="m")
plt.show()
```

```python
import plotly.express as px
# Create an interactive widget to select features.
from ipywidgets import widgets
from IPython.display import display, clear_output

# Create the dropdown widget for feature selection
dropdown = widgets.Dropdown(
    options=features,
    value='tempo',
    description='Feature:',
    disabled=False,
)

# Placeholder for the Plotly figure
figure_placeholder = widgets.Output()

def update_chart(change):
    with figure_placeholder:
        clear_output(wait=False)  # Clear the previous chart
        # Create a new figure with the selected feature
        fig = px.histogram(df_union, x=change.new, color="track_genre_top", barmode="overlay",
                           facet_col="track_genre_top", facet_col_wrap=4)
        fig.update_layout(height=400,showlegend=False)  # Set the desired new height for your chart
        fig.show()

# When the dropdown value changes, update the chart with the new feature
dropdown.observe(update_chart, names='value')

# Display the dropdown
display(dropdown)

# Display the Plotly figure output placeholder
display(figure_placeholder)
```
