```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
## Genre
genre_df = pd.read_csv(GENRE_FILE)
genre_df.head()
```

```python
G = nx.DiGraph()

for index, row in genre_df.iterrows():
    G.add_node(row['genre_id'], label=row['title'])
    
for index, row in genre_df.iterrows():
    if pd.notna(row['parent']):
        G.add_edge(row['parent'], row['genre_id'])
        
        
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 6))

nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color='skyblue')
plt.title("Genre Hierarchical Tree")
plt.show()
```

```python
## it is too complicated.
## Just get leaf noeds.

# Identify leaf nodes
leaf_nodes = [node for node, outdegree in G.out_degree() if outdegree == 0]

# Extract information about each leaf node, its parent, and top level genre
leaf_info = []
for leaf in leaf_nodes:
    parents = list(G.predecessors(leaf))
    parent = parents[0] if parents else None  # Check if node has a parent
    
    leaf_info_dict = {
        'leaf_id': leaf,
        'leaf_name': genre_df[genre_df['genre_id'] == leaf]['title'].values[0],
        'parent_id': parent,
        'top_level_genre_id': genre_df[genre_df['genre_id'] == leaf]['top_level'].values[0]
    }
    
    # Add parent name only if parent exists (not a root node)
    if parent:
        leaf_info_dict['parent_name'] = genre_df[genre_df['genre_id'] == parent]['title'].values[0]
    else:
        leaf_info_dict['parent_name'] = None

    leaf_info_dict['top_level_name'] = genre_df[genre_df['genre_id'] == leaf_info_dict['top_level_genre_id']]['title'].values[0]

    leaf_info.append(leaf_info_dict)

leaf_df = pd.DataFrame(leaf_info)





```
